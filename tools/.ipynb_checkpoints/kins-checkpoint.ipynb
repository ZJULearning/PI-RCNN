```python
import os
import numpy as np
import json
import pycocotools.mask as maskUtils
import matplotlib.pyplot as plt

class KINSDataset():
    def __init__(self, raw_data_dir='/workspace/data/Kitti/raw/kins', split='train', class_to_id=None, classes=['car', 'pedestrian', 'cyclist']):
        self.kins = json.load(open(os.path.join(raw_data_dir, 'update_%s_2020.json' % split), 'r'))
        self.image_map = {}
        for image in self.kins['images']:
            self.image_map[image['id']] = image
        self.id_to_category = {}
        for category in self.kins['categories']:
            self.id_to_category[category['id']] = category['name']
        self.label_map = {}
        for i in range(len(self.kins['annotations'])):
            anno = self.kins['annotations'][i]
            anno['category'] = self.id_to_category[anno['category_id']]
            anno['image'] = self.image_map[anno['image_id']]
            image_name = os.path.splitext(anno['image']['file_name'])[0]
            if image_name not in self.label_map:
                self.label_map[image_name] = [anno]
            else:
                self.label_map[image_name].append(anno)
        if class_to_id is None:
            self.class_to_id = {
                'car': 1,
                'pedestrian': 2,
                'cyclist': 3,
            }
        else:
            self.class_to_id = class_to_id
        self.classes = classes

    def __getitem__(self, sample_name):
        assert sample_name in self.label_map
        
        annos = self.label_map[sample_name]
        img_h, img_w = annos[0]['image']['height'], annos[0]['image']['width']
        sem_mask = np.ones((img_h, img_w), dtype=np.int32) * -1
        sorted_annos = sorted(annos, key=lambda x: x['oco_id'], reverse=True)
        for i, anno in enumerate(sorted_annos):
            height = anno['image']['height']
            width = anno['image']['width']
            category_id = anno['category_id']
            if not self.id_to_category[category_id].lower() in self.classes:
                continue
            category_id = self.class_to_id[self.id_to_category[category_id].lower()]
            _insta_mask = maskUtils.decode(maskUtils.frPyObjects(anno['i_segm'], height, width))[:, :, 0]
            sem_mask[sem_mask == -1] += (_insta_mask > 0).astype(np.int32)[sem_mask == -1] * category_id
        return sem_mask

```

```python
dataset = KINSDataset()
```

```python
sample_name = list(dataset.label_map.keys())[9]
sample = dataset[sample_name]
print(np.unique(sample, return_counts=True))

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(311)
ax.imshow(sample)

import cv2
image = cv2.cvtColor(cv2.imread(os.path.join('/workspace/data/Kitti/object/training/image_2/%s.png' % sample_name)), cv2.COLOR_BGR2RGB)
ax = fig.add_subplot(312)
ax.imshow(image)

print(np.unique(sample.astype(np.uint8), return_counts=True))
sample_rgb = cv2.applyColorMap(sample.astype(np.uint8), cv2.COLORMAP_JET)
ax = fig.add_subplot(313)
ax.imshow(cv2.addWeighted(image, 0.8, sample_rgb, 0.8, gamma=0))
```
