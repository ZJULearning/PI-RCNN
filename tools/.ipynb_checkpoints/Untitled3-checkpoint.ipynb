```python
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
```

```python
sample_list = [s.strip() for s in open('../data/kitti/ImageSets/train.txt').readlines()]
```

```python
sample_id = sample_list[90]
image = cv2.imread(os.path.join('../data/kitti/training/image_2/', f'{sample_id}.png'))
pred1 = np.load(os.path.join('../output/kitti_models/pi_rcnn_seg_pretrain_convert_label/default/eval/epoch_200/train/default/final_result/data', f'{sample_id}.npy'))

pred2 = np.load(os.path.join('../output/kitti_models/pi_rcnn_seg_pretrain_convert_label/default/eval/epoch_200/train/default/final_result/data', f'{sample_id}.npy'))


image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(311)
ax.imshow(image)

print(np.max(pred1), np.min(pred1))
pred1_rgb = cv2.applyColorMap((pred1[1] * 255).astype(np.uint8), cv2.COLORMAP_JET)
pred1_rgb = cv2.cvtColor(pred1_rgb, cv2.COLOR_BGR2RGB)
ax = fig.add_subplot(312)
ax.imshow(cv2.addWeighted(image, 0.8, pred1_rgb, 0.8, gamma=0))


print(np.max(pred2), np.min(pred2))
pred2_rgb = cv2.applyColorMap((pred2[1] * 255).astype(np.uint8), cv2.COLORMAP_JET)
pred2_rgb = cv2.cvtColor(pred2_rgb, cv2.COLOR_BGR2RGB)
ax = fig.add_subplot(313)
ax.imshow(cv2.addWeighted(image, 0.8, pred2_rgb, 0.8, gamma=0))

```
