```python
from pcdet.datasets.kitti.kitti_seg_dataset import KittiSegDataset
import matplotlib.pyplot as plt
from pcdet.config import cfg, cfg_from_yaml_file
import numpy as np
import os

%load_ext autoreload
%autoreload 2
```

```python
def parse_config(cfg_file):
    cfg_from_yaml_file(cfg_file, cfg)
    return cfg

cfg = parse_config('cfgs/kitti_models/pi_rcnn_seg_pretrain_convert_label_kins.yaml')
```

```python
dataset = KittiSegDataset(cfg.DATA_CONFIG, class_names=cfg.CLASS_NAMES)
```

```python
sample = dataset[3390]

image, label = sample['image'], sample['image_seg_label']
print(label.shape)
print(sample['sample_id'])
print(label.dtype)
print(np.unique(label, return_counts=True))

label[label == 255] = 0
image = image.astype(np.uint8)

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(311)
ax.imshow(label)

import cv2
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
ax = fig.add_subplot(312)
ax.imshow(image)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
ax = fig.add_subplot(313)
label_rgb = cv2.cvtColor(cv2.applyColorMap((label * 80).astype(np.uint8), cv2.COLORMAP_JET), cv2.COLOR_BGR2RGB)
ax.imshow(cv2.addWeighted(image, 0.8, label_rgb, 0.8, gamma=0))
```
